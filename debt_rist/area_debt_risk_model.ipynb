```python
#数据[0,1]标准化
def MaxMinNormalization(x):
    """[0,1]normalization"""
    x=(x-np.min(x))/(np.max(x)-np.min(x))
    return x

def MinMaxNormalization(x):
    """[0,1]Backward normalization"""
    x=(np.max(x)-x)/(np.max(x)-np.min(x))
    return x
```
